```python
ac_attributes.loc[:,"PM_Kisan_merged_amount_paid"] = ac_attributes["PM_Kisan_merged_amount_paid"].apply(lambda x: float(x.replace(',','')) if isinstance(x, str) else x)
```

```python
ac_attributes.loc[:,"PM_Kisan_merged_amount_due"] = ac_attributes["PM_Kisan_merged_amount_due"].apply(lambda x: float(x.replace(',','')) if isinstance(x, str) else x)
```

```python
ac_df = ac_attributes[['AC_NAME','AC_NO','PM_Kisan_merged_reg_count', 'PM_Kisan_merged_full_pay_count','PM_Kisan_merged_part_pay_count','PM_Kisan_merged_no_pay_count','PM_Kisan_merged_extra_pay_count','PM_Kisan_merged_no_install_paid','PM_Kisan_merged_no_install_due','PM_Kisan_merged_amount_paid','PM_Kisan_merged_amount_due']]
```

```python
ac_df
```

```python
AC_wise_PM_Kisan = ac_df.groupby(['AC_NAME','AC_NO']).sum(['PM_Kisan_merged_amount_due'])
```
